# Lectura del conjunto de datos y preparación del DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('movie_metadata.csv', encoding='UTF-8')
df
```

```python
# Se añade una columna llamada Id que es igual al índice
df['Id'] = df.index

# Se mueve dicha columna al inicio del DataFrame
df = df[['Id'] + [col for col in df.columns if col != 'Id']]

df
```

```python
# Se visualiza la columna genres y plot_keywords, que son las que interesan.
df[['genres', 'plot_keywords']]
```

```python
# Se reemplazan las | por espacios
df['genres'] = df['genres'].str.replace('|', ' ')
df['plot_keywords'] = df['plot_keywords'].str.replace('|', ' ')
df[['genres', 'plot_keywords']]
```

```python
# Se crea una columna llamada texto en la que se unan las columnas genres y plot_keywords

df['texto'] = df[['genres', 'plot_keywords']].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df['texto']
```

# Aplicación del método TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
```

```python
# Se instancia el método TfidfVectorizer y se le indica la cantidad máxima de tokens

tfidf = TfidfVectorizer(max_features=2000)
```

```python
# Se va a crear la matriz de vectores aplicando el método tf-idf a la variable texto que se ha creado

X = tfidf.fit_transform(df['texto'])
X
```

```python
# Se va a crear un DataFrame con la variable movie_title como índice

peliculas = pd.Series(df.index, index=df['movie_title'])
peliculas
```

```python
# Imprime los índices disponibles para verificar cómo se almacenan
print(peliculas.index)
```

```python
# Elimina espacios alrededor de los nombres de las películas en el índice
peliculas.index = peliculas.index.str.strip()
```

```python
# Se crea una variable llamada indice que va a coger una película del DataFrame peliculas

indice = peliculas['The Dark Knight Rises']
indice
```

```python
# Se crea una variable llamada consulta accediendo a la posicion guardada en la variable indice
consulta = X[indice]
consulta
```

Como se ve en el output anterior, la variable consulta es un vector de 1x2000 (es decir, 2000 dimensiones)

```python
# Se imprime el vector consulta para verlo con más detalle. Tan sñolo va a mostrar los valores donde el valor es diferente de 0
print(consulta)
```

```python
consulta.toarray()
```

```python
# Se va a crear la variable similitud, lo que va a ser un arreglo de similitud. Esto va a devolver la similitud entre la pelicula
# seleccionada (The Dark Knight Rises) con el resto de peliculas guardadas en el DataFrame X
similitud = cosine_similarity(consulta, X)
```

```python
# Se muestra la similitud. Cuanto más valor, más parecida es la pelicula del df con la seleccionada
similitud
```

```python
# Si se ve la similitud frente a la misma película, va a ser 1:
similitud[0][3]
```

```python
# Se va a transformar la variable similitud en un array
similitud = similitud.flatten()
similitud[3]
```

```python
# Se van a graficar las solicitudes

import matplotlib.pyplot as plt

plt.plot(similitud)
```

```python
# Se va a ordenar la similitud
plt.plot(similitud[(-similitud).argsort()])
```

# Recomendador de película

Se van a recomendar las 10 más parecidas a la película seleccionada: The Dark Knight Rises

```python
# Se crea una vairable llamada recomendación que recorre la variable similitud desde la posición 1 hasta la 11, así se evita que
# recomiende la misma película. Esto devuelve las posiciones de cada película

recomendacion = (-similitud).argsort()[1:11]
recomendacion
```

```python
# Se van a mostrar los títulos de las 10 películas más parecidas a The Dark Knight Rises

df['movie_title'].iloc[recomendacion]
```
